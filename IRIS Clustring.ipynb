```python

import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score
```

```python
iris = datasets.load_iris()
```

```python
X = iris.data
y = iris.target
```

```python
MM = MinMaxScaler()
x = MM.fit_transform(X)
```

```python
y[y==0] = 10
y[y == 1] = 0
y[y == 10] = 1
```

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap='gist_rainbow')
plt.xlabel('Spea1 Length', fontsize=18)
plt.ylabel('Sepal Width', fontsize=18)
```

# Kmean

```python
km = KMeans(n_clusters = 3, n_jobs = 4, random_state=21)
km.fit(X)
```

```python
centers = km.cluster_centers_
print(centers)
```

```python
y_pred_kmean = km.predict(X)
```

```python
accuracy_score(y, y_pred_kmean)
```

```python
new_labels = km.labels_
# Plot the identified clusters and compare with the answers
fig, axes = plt.subplots(1, 2, figsize=(16,8))
axes[0].scatter(X[:, 0], X[:, 1], c=y, cmap='gist_rainbow',
edgecolor='k', s=150)
axes[1].scatter(X[:, 0], X[:, 1], c=new_labels, cmap='jet',
edgecolor='k', s=150)
axes[0].set_xlabel('Sepal length', fontsize=18)
axes[0].set_ylabel('Sepal width', fontsize=18)
axes[1].set_xlabel('Sepal length', fontsize=18)
axes[1].set_ylabel('Sepal width', fontsize=18)
axes[0].tick_params(direction='in', length=10, width=5, colors='k', labelsize=20)
axes[1].tick_params(direction='in', length=10, width=5, colors='k', labelsize=20)
axes[0].set_title('Actual', fontsize=18)
axes[1].set_title('Predicted', fontsize=18)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_pred_kmean, cmap="plasma")
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```

# Agglomerative 

```python
from sklearn import datasets, cluster
```

```python
X = datasets.load_iris().data
```

```python
cluster = cluster.AgglomerativeClustering(n_clusters=3, linkage='ward')
```

```python
y_pred_agg = cluster.fit_predict(X)
```

```python
accuracy_score(y, y_pred_agg)
```

```python
from scipy.cluster.hierarchy import dendrogram, ward, single
```

```python
linkage_matrix = ward(X)
```

```python
dendrogram(linkage_matrix)
plt.show()
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_pred_agg, cmap="plasma")
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```

# Density

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=0.123, min_samples = 2)
y_pred_density = dbscan.fit_predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_pred_density, cmap="plasma")
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```

```python
accuracy_score(y, y_pred_density)
```
